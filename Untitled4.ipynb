```python
from os import X_OK
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/shopping_trends.csv')
df.head()
x=np.arange(0.50,np.pi*4,0.60)
y1= np.sin(x)
y2= np.cos(x)
plt.plot(x,y1,color='b',linestyle='-.',label='sin(x)')
plt.plot(x,y2, 'y--', label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('sin-cos graph')
plt.legend(loc='lower right')

df.head()
```
